# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()

# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
#view keys 
Base.classes.keys()
```

```python
#save table variables
measurement = Base.classes.measurement
station = Base.classes.station
```

```python
#inspect columns
#first setup inspector with engine
inspector = inspect(engine)
columns = inspector.get_columns('measurement')
for header in columns:
    print(header['name'], header['type'])
```

```python
#inspect station column
inspector = inspect(engine)
columns = inspector.get_columns('measurement')
for header in columns:
    print(header['name'], header['type'])
```

```python
# Create our session (link) from Python to the DB
#create session 
session = Session(engine)
```

# Exploratory Climate Analysis

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results
#first, get the latest date
latest_date = session.query(measurement.date).order_by(measurement.date.desc()).first()
latest_date = latest_date[0]
print(latest_date)
```

```python
#subtract 365 days to calculate 1 year ago from the last data point in the database
one_year = dt.datetime.strptime(latest_date, "%Y-%m-%d") - dt.timedelta(days=365)
```

```python
# Query data from a year ago
precip_query = session.query(measurement.date, measurement.prcp).filter(measurement.date >= one_year).all()
precip_query
```

```python
#create new data frame for query data
df = pd.DataFrame(precip_query,columns=['date', 'precipitation'])
```

```python
# Sort the dataframe by date and set index
df.set_index('date', inplace = True)
df.sort_values(by='date',ascending= True) 
```

```python
#plot data on bar graph
df.plot(title="2016-2017 Precipitation in Hawaii",rot=45)
plt.legend(loc='upper left')
plt.ylabel("Inches")
plt.xlabel("Date")
plt.savefig("Images/Precipitation.png")
plt.show()
```

```python
#Stat table for data set 
df.describe()
```

```python
# query station count
session.query(station.station).count()
```

```python
# What are the most active stations? (i.e. what stations have the most rows)?
# List the stations and the counts in descending order.
#query most active station
station_active = session.query(measurement.station, func.count(measurement.id)).\
            group_by(measurement.station).\
            order_by(func.count(measurement.id).desc()).all()
station_active
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature of the most active station?
#manually put in statipn ID 
#query and use caculation functions
station_temp = session.query(func.min(measurement.tobs), func.max(measurement.tobs),
                                  func.avg(measurement.tobs)).\
                filter(measurement.station == 'USC00519281').all()
degree_sign = u"\N{DEGREE SIGN}"
print(f"Most Active Station is USC00519281")
print(f"Lowest Temperature: {station_temp[0][0]}{degree_sign} F")
print(f"Highest Temperature: {station_temp[0][1]}{degree_sign} F")
print(f"Average Temperature: {round(station_temp[0][2], 2)}{degree_sign} F")
```

```python
# Choose the station with the highest number of temperature observations.
station_high = session.query(measurement.station, func.count(measurement.tobs)).\
                group_by(measurement.station).\
                order_by(func.count(measurement.station).desc()).first()
station_high
```

```python
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
station_temp = session.query(measurement.station, measurement.tobs).\
                filter(measurement.station == station_high[0]).\
                filter(measurement.date >= one_year).all()
station_temp
```

```python
# Plot histagram
temp_df = pd.DataFrame(station_temp, columns = ["station", "tobs"])
temp_df.hist(column="tobs", bins = 12)
plt.ylabel("Frequency")
plt.legend(["Temperature (° F)"], loc = "upper left")
plt.title("Station USC00519281 Temperature over a Year Period")
plt.savefig("Images/Temp.png")
plt.show()
```

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```
